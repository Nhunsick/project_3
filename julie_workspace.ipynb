```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, RocCurveDisplay, ConfusionMatrixDisplay, classification_report
```

```python
# columns = ['dem_win', 'state', 'pres_16_mov', 'pres_12_mov', 'fips', 'median_hh_inc', 'clf_unemploy_pct', 
#         'lesshs_pct', 'lesscollege_pct', 'lesshs_whites_pct', 'lesscollege_whites_pct', 'rural_pct', 'age29andunder_pct',
#           'age65andolder_pct', 'female_pct', 'total_population','cvap', 
#            'white_pct', 'black_pct', 'hispanic_pct', 'nonwhite_pct', 'foreignborn_pct']

X = merged_df_no_sen.drop(columns=['dem_win', 'state', 'fips'], axis=1)
# X = merged_df_no_sen[['pres_16_mov']]
y = merged_df_no_sen['dem_win']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=34567)
```

```python
tree_clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=1234)
tree_clf.fit(X_train, y_train)
```

```python
f, ax =  plt.subplots(figsize=(20,20))
plot_tree(tree_clf, ax=ax);
```

```python
tree_clf.score(X_train, y_train)
```

```python
y_pred = tree_clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
acc
```

```python
ConfusionMatrixDisplay.from_estimator(tree_clf, X_test, y_test);
```
