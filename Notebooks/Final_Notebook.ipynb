## MIT data exclusions

The first group of features from the election_results dataframe that is not present in our merged dataframe is the individual candidate vote totals for past elections (ex. trump16, obama12, etc...). Instead what we have in our merged dataframe is a margin of victory column for each of the elections that were represented by the individual candidate vote totals. The reason for this is that the magnitude of a candidates vote total changes with different population counties and we wanted to control for that. Furthermore, we only included districts where a major two party candidate won each of the elections that we used as our predictors. This is beacuse the problem becomes much more complex when you add in third party candidates, and in America's electoral system third party candidates winning is seen as a major exception se we did not see the need to inlcude them. 
Finally, we decided to not include senate or gubenatorial data in our modeling as there were many states that we did not have the most recent data for for these races as they do not happen on a nation wide regular cycle like house and presidential elections do.
