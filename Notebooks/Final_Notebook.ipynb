## Data Exclusions

All of the cleaning mentioned here can be seen in the EDA.ipynb notebook with code comments on our process.

The first group of features from our data that is not present in our merged dataframe is the individual candidate vote totals for elections (ex. trump16, obama12, as well as vote1 and vote2 for our target). Instead what we have in our merged dataframe is a margin of victory column for each of the elections that were represented by the individual candidate vote totals. The reason for this is that the magnitude of a candidates vote total changes with different population counties and we wanted to control for that. All of the MOV statistics go from a scale of -1 to 1 where -1 is republicans winning the vote by 100% and a 1 represents the democrat winning the county by 100%, the way we grouped these MOV categories can be seen below.


```python
Image("../images/mov_scale.png")
```

Furthermore, we only included districts where a major two party candidate won each of the elections that we used as our predictors. This is beacuse the problem becomes much more complex when you add in third party candidates, and in America's electoral system third party candidates winning is seen as a major exception se we did not see the need to inlcude them. 

Finally, we decided to not include senate or gubenatorial data in our modeling as there were many states that we did not have the most recent data for for these races as they do not happen on a nation wide regular cycle like house and presidential elections do.  

Other exclusions from the county data dataset include:
1. white_pct, as this is perfectly inversely multicollinear with nonwhite_pct
2. state, fips, name, as well as county from our election results dataframe, these were excluded as they were merely indexing data that would not be needed for our machine learning models
3. cvap, which stands for citizen voting age population, this was transformed into cvap_pct which is the citizen voting age population divided by the total population, again this is just another standardizing procedure we needed to run our model
4. ruralurban_cc, this is a sliding scale which dictates how rural a county classifies as, not needed as we already have rural_pct as a measure or how rural a county is
5. All of the vote categories over 2 in our election results dataframe, these were votes for third party candidates which we did not need

## Imputation

There was one data imputation step that we needed to include in our cleaning and it has to do with uncontested elections. If you look at the data for congressional elections in 2016 you will see that there are multiple rows that have zeros for one party's vote total. This is not an error but a signifier that one of the parties decided not to run a candidate in a given district which led to an uncontested election. We could not leave it as is so instead we wanted to find a good predictor of what the margin of victory would have been for that district if there had been a contested election in that county in 2016. What we decided to do was to assign the margin of victory for these elections to be the margin of victory for the 2016 presidential election. The reason we did this is because in presidential election years such as 2016 people are much more likely to vote straight ticket ballots, and the president to house voting behavior is more similar than either senate to house or governor to house behavior.
