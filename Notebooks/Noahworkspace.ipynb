```python
#dont need state or fips anymore, not essential to our modeling
drop_for_model = ['state', 'fips']
merged_df = merged_df.drop(columns = drop_for_model)
merged_df_no_sen = merged_df_no_sen.drop(columns = drop_for_model)
merged_df
```

```python
#making a base logistic regression model, this code generates a train test split for machine learning
from sklearn.model_selection import train_test_split
#demwin boolean is our target
X = merged_df_no_sen.drop(columns = 'dem_win')
y = merged_df_no_sen['dem_win']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 21)
```

```python
#performing cross validation to make sure our model is consistent and has tolerable variance and bias
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression

def cross_validation(X_train, y_train, c, num_split = 10):
    
    X_train = X_train.values
    y_train = y_train.values
    
    score_train_list = []
    score_val_list = []
    
    for train_index, valid_index in KFold(n_splits = num_split).split(X_train):
        X_train_fold, X_val_fold = X_train[train_index], X_train[valid_index]
        y_train_fold, y_val_fold = y_train[train_index], y_train[valid_index]
        scaler = MinMaxScaler()
        X_tf_sc = scaler.fit_transform(X_train_fold)
        X_vld_sc = scaler.transform(X_val_fold)

        logreg = LogisticRegression(fit_intercept=False, solver='liblinear', C = c)
        logreg.fit(X_tf_sc, y_train_fold)

        accuracy_train = logreg.score(X_tf_sc, y_train_fold)
        accuracy_val = logreg.score(X_vld_sc, y_val_fold)
        score_val_list.append(accuracy_val)
        score_train_list.append(accuracy_train)
    return {'c': c, 'train': np.mean(score_train_list), 'validation': np.mean(score_val_list)}
```

```python
#this iterates through different magnitudes of c, our regularization coefficent, to find which results in our best outcomes
C_list = [1e-4,1e-3, 1e-2, 1e-1, 1, 10, 100, 1e3, 1e4]
crossval_df = pd.DataFrame(columns = ['c', 'train', 'validation'])
for c in C_list:
    crossval_df = crossval_df._append(cross_validation(X_train, y_train, c, 10), ignore_index = True)
```

```python
#now that we have cross validated our model we can acutally use it, made a train test split and performed the logistic regression, this will
#serve as our baseline model
fulltrain_scaler = MinMaxScaler()
#using minmaxscaler over standard scaler because our data is not normally distributed
X_train_scaled = fulltrain_scaler.fit_transform(X_train)
X_test_scaled = fulltrain_scaler.transform(X_test)
logreg_final = LogisticRegression(fit_intercept = False, solver = 'liblinear', C = 10)
logreg_final.fit(X_train_scaled, y_train)
```

```python
#checking the model's performance
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score
y_hat_test = logreg_final.predict(X_test_scaled)
y_hat_train = logreg_final.predict(X_train_scaled)
print('Training Precision: ', precision_score(y_train, y_hat_train))
print('Testing Precision: ', precision_score(y_test, y_hat_test))
print('\n\n')
print('Training Recall: ', recall_score(y_train, y_hat_train))
print('Testing Recall: ', recall_score(y_test, y_hat_test))
print('\n\n')
print('Training Accuracy: ', accuracy_score(y_train, y_hat_train))
print('Testing Accuracy: ', accuracy_score(y_test, y_hat_test))
print('\n\n')
print('Training F1-Score: ', f1_score(y_train, y_hat_train))
print('Testing F1-Score: ', f1_score(y_test, y_hat_test))
```

```python
#running an l1 penalty regression to see if it causes any difference with the possible multicollinearity we have with the features mentioned
#above
def cross_validation_l1(X_train, y_train, c, num_split = 10):
    
    X_train = X_train.values
    y_train = y_train.values
    
    score_train_list = []
    score_val_list = []
    
    for train_index, valid_index in KFold(n_splits = num_split).split(X_train):
        X_train_fold, X_val_fold = X_train[train_index], X_train[valid_index]
        y_train_fold, y_val_fold = y_train[train_index], y_train[valid_index]
        scaler = MinMaxScaler()
        X_tf_sc = scaler.fit_transform(X_train_fold)
        X_vld_sc = scaler.transform(X_val_fold)

        logreg = LogisticRegression(fit_intercept=False, penalty = 'l1', solver='liblinear', C = c)
        logreg.fit(X_tf_sc, y_train_fold)

        accuracy_train = logreg.score(X_tf_sc, y_train_fold)
        accuracy_val = logreg.score(X_vld_sc, y_val_fold)
        score_val_list.append(accuracy_val)
        score_train_list.append(accuracy_train)
    return {'c': c, 'train': np.mean(score_train_list), 'validation': np.mean(score_val_list)}
```

```python
#this is making a new column that bins the margins of victory based on how large they are, 1 is a major republican victory and 6 is a major
#democrat victory
mov_df['mov_scale'] = 0
mov_df.loc[(mov_df['target_mov'] >= -1) & (mov_df['target_mov'] <= -.20), 'mov_scale'] = 1
mov_df.loc[(mov_df['target_mov'] > -.20) & (mov_df['target_mov'] <= -.10), 'mov_scale'] = 2
mov_df.loc[(mov_df['target_mov'] > -.10) & (mov_df['target_mov'] <= 0), 'mov_scale'] = 3
mov_df.loc[(mov_df['target_mov'] > 0) & (mov_df['target_mov'] <= .10), 'mov_scale'] = 4
mov_df.loc[(mov_df['target_mov'] > .10) & (mov_df['target_mov'] <= .20), 'mov_scale'] = 5
mov_df.loc[(mov_df['target_mov'] > .20) & (mov_df['target_mov'] <= 1), 'mov_scale'] = 6
```
