```python
#importing all necessary libraries
import pandas as pd
import numpy as np
```

```python
#uploading the datasets
county_data = pd.read_csv('data/election-context-2018.csv')
election_results = pd.read_csv('data/2018_4_0_2.csv', skiprows = [0])
```

```python
#merging the two datasets
merged_df = county_data.merge(election_results, left_on = "fips", right_on = 'fips')
merged_df.head()
```

```python
list_to_drop = []
for n in range(3,23):
    list_to_drop.append("vote" + str(n))
list_to_drop
```

```python
#dropping columns
merged_df = merged_df.drop(columns = list_to_drop)
merged_df.info()
```

```python
#creating new columns to account for the total votes for each race 
merged_df['pres_16_total_votes'] = merged_df['trump16'] + merged_df['clinton16'] + merged_df['otherpres16']
merged_df['pres_12_total_votes'] = merged_df['romney12'] + merged_df['obama12'] + merged_df['otherpres12']
merged_df['sen_16_total_votes'] = merged_df['repsen16'] + merged_df['demsen16'] + merged_df['othersen16']
merged_df['house_16_total_votes'] = merged_df['rephouse16'] + merged_df['demhouse16'] + merged_df['otherhouse16']
merged_df
```

```python
#dropping gov because of missing null values 
list_to_drop_2 = ['demgov16', 'repgov16', 'othergov16', 'demgov14', 'repgov14', 'othergov14']
merged_df = merged_df.drop(columns = list_to_drop_2)
merged_df.info()
```

```python
#dropping more columns
list_to_drop_3 = ['name', 'county', 'type']
merged_df = merged_df.drop(columns = list_to_drop_3)
merged_df.info()
```

```python
#seeing if there were any third-party candidates with more votes than dems and reps for 2016 presidential race
third_party_df_pres_16 = merged_df.loc[(merged_df['otherpres16'] >= merged_df['trump16']) & (merged_df['otherpres16'] >= merged_df['clinton16'])]
merged_df = merged_df.loc[~merged_df['fips'].isin(third_party_df_pres_16['fips'])]
merged_df
```

```python
#seeing if there were any third-party candidates with more votes than dems and reps for 2012 presidential race
third_party_df_pres_12 = merged_df.loc[(merged_df['otherpres12'] >= merged_df['romney12']) & (merged_df['otherpres12'] >= merged_df['obama12'])]
merged_df = merged_df.loc[~merged_df['fips'].isin(third_party_df_pres_12['fips'])]
merged_df
```

```python
#seeing if there were any third-party candidates with more votes than dems and reps for senate race
third_party_df_sen_16 = merged_df.loc[(merged_df['othersen16'] >= merged_df['repsen16']) & (merged_df['othersen16'] >= merged_df['demsen16'])]
merged_df = merged_df.loc[~merged_df['fips'].isin(third_party_df_sen_16['fips'])]
merged_df
```

```python
#seeing if there were any third-party candidates with more votes than dems and reps for house race 
third_party_df_house_16 = merged_df.loc[(merged_df['otherhouse16'] >= merged_df['rephouse16']) & (merged_df['otherhouse16'] >= merged_df['demhouse16'])]
merged_df = merged_df.loc[~merged_df['fips'].isin(third_party_df_house_16['fips'])]
merged_df
```

```python
#creating new columns to get the margin of victory for the 2016 presidential election 
merged_df['pres_16_vic_size'] = merged_df['clinton16'] - merged_df['trump16']
merged_df['pres_16_mov'] = merged_df['pres_16_vic_size'] / merged_df['pres_16_total_votes']
merged_df
```

```python
#creating new columns to get the margin of victory for 2012 presidential election 
merged_df['pres_12_vic_size'] = merged_df['obama12'] - merged_df['romney12']
merged_df['pres_12_mov'] = merged_df['pres_12_vic_size'] / merged_df['pres_12_total_votes']
merged_df
```

```python
#creating new columns to get the margin of victory for the senate
merged_df['sen_16_vic_size'] = merged_df['demsen16'] - merged_df['repsen16']
merged_df['sen_16_mov'] = merged_df['sen_16_vic_size'] / merged_df['sen_16_total_votes']
merged_df
```

```python
#creating new columns to get the margin of victory for the house
merged_df['house_16_vic_size'] = merged_df['demhouse16'] - merged_df['rephouse16']
merged_df['house_16_mov'] = merged_df['house_16_vic_size'] / merged_df['house_16_total_votes']
merged_df
```

```python
#dropping third parties, focusing on dems and reps
other_party_drop_list = ['otherpres12', 'otherpres16', 'othersen16', 'otherhouse16']
merged_df = merged_df.drop(columns = other_party_drop_list)
merged_df
```

```python
#cols_to_drop_for_sen = ['demsen16', 'repsen16', 'sen_16_total_votes', 'sen_16_vic_size', 'sen_16_mov']
#merged_df_no_sen = merged_df.drop(columns = cols_to_drop_for_sen)
```

```python
#detects nans or zeros for the uncontested house elections 
dem_house_nan_df = merged_df.loc[(merged_df['demhouse16'].isna()) | (merged_df['demhouse16'] == 0) | (merged_df['rephouse16'].isna()) | (merged_df['rephouse16'] == 0)]

#take all the values with nans/zeros in those rows and replace them with the values from 'pres_16_mov'
merged_df.loc[merged_df['fips'].isin(dem_house_nan_df['fips']), 'house_16_mov'] = merged_df['pres_16_mov']

merged_df
```

```python
#dropping more columns
vote_size_drop = ['pres_12_total_votes', 'pres_16_total_votes', 'sen_16_total_votes', 'house_16_total_votes', 'pres_12_vic_size', 'pres_16_vic_size', 'sen_16_vic_size', 'house_16_vic_size']
merged_df = merged_df.drop(columns = vote_size_drop)
merged_df
```

```python
#dropping candidates columns
candidate_drop = ['trump16', 'clinton16', 'obama12', 'romney12', 'demsen16', 'repsen16', 'demhouse16', 'rephouse16']
merged_df = merged_df.drop(columns = candidate_drop)
merged_df
```

```python
#creating new column called "dem_win"
merged_df['dem_win'] = np.where(merged_df['vote1'] > merged_df['vote2'], 1, 0)
merged_df
```

```python
final_drop_mov = ['vote1', 'vote2', 'totalvote', 'ruralurban_cc', 'white_pct', 'target_vic_size', 'dem_win', 'sen_16_mov']
mov_df = merged_df.copy()
mov_df['target_vic_size'] = mov_df['vote1'] - mov_df['vote2']
mov_df['target_mov'] = mov_df['target_vic_size'] / mov_df['totalvote']
mov_df = mov_df.drop(columns = final_drop_mov)
```

```python
#dropping columns
final_drop = ['vote1', 'vote2', 'totalvote', 'ruralurban_cc', 'white_pct']
merged_df = merged_df.drop(columns = final_drop)
merged_df.head()
```

```python
#detect and remove the one nan value in 'cvap'
merged_df_nan_here = merged_df[merged_df['cvap'].isna()]
merged_df = merged_df.loc[~merged_df['fips'].isin(merged_df_nan_here['fips'])]
```

```python
#creating a dataframe without the senate info
merged_df_no_sen = merged_df.drop(columns = 'sen_16_mov')
merged_df_no_sen
```

```python
#creating new column to get the percentage of voters 18+
merged_df_no_sen['cvap_pct'] =  merged_df_no_sen['cvap']/merged_df_no_sen['total_population']
merged_df_no_sen
```
